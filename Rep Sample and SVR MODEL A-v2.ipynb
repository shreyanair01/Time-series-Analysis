# Modelling

```python
X = train_A.copy()
```

```python
X = X.drop(['StoreType','Promo2','Promo2SinceWeek','Promo2SinceYear','PromoInterval','Sales','CompetitionOpenSinceMonth','CompetitionOpenSinceYear'], axis =1)

X.columns
```

```python
X.columns
```

```python
# from sklearn.preprocessing import LabelEncoder, OneHotEncoder
# le_state_holiday = LabelEncoder()
# X['StateHoliday'] = le_state_holiday.fit_transform(X['StateHoliday'])
# # onc_state_holiday = OneHotEncoder()


# list(le_state_holiday.classes_)
#['0', 'a']
```

NOT VALID NOW as 4 classes are available
Since there is only one type of holiday, it does not make sense to keep using label encoder, we can simply convert all state holiday to 1 (as using label encoder converts all a to 0, but 0 indicates not holiday)

```python
map_holiday = {'a':1,'b':2, 'c':3,'d':4, '0':0}

X['StateHoliday'] = X['StateHoliday'].map(map_holiday)

set(X['StateHoliday'].values)
```

```python
X.groupby(['StateHoliday']).count()
```

```python
X = pd.concat([X,pd.get_dummies(X['StateHoliday'],prefix=['StateHoliday_type'],drop_first=True)],axis=1)

X = X.drop(['StateHoliday'],axis =1)
```

```python
X.head()
```

```python
# from sklearn.preprocessing import LabelEncoder, OneHotEncoder
# onc_state_holiday = OneHotEncoder()
# X['StateHoliday'] = onc_state_holiday.fit_transform(X['StateHoliday'].values.reshape(-1,1))
# # encoded_stateholiday = [i[0] for i in encoded_stateholiday]
# encoded_stateholiday
```



Before transforming assortment, we check which types are available

```python
X.groupby(['Assortment']).count()
```

since only 2 labels are present, we can use maps again

```python
map_assortment = {'a':0,'c':1}

X['Assortment'] = X['Assortment'].map(map_assortment) 

set(X['Assortment'].values)

y = train_A['Sales'].values.reshape(-1,1)

# store date index before scaling
train_size = round(0.80*len(X))
X_train_dates = X.index[:train_size]
X_test_dates = X.index[train_size:]
y_train_dates = X.index[:train_size]
y_test_dates = X.index[train_size:]

from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X=sc.fit_transform(X)
sc_y = StandardScaler()
y = sc_y.fit_transform(y)
```

```python
len(X)==len(y)
```

```python
X_train = X[:train_size,:].copy()
X_test = X[train_size:,:].copy()
y_train = y[:train_size].copy()
y_test = y[train_size:].copy()
```

```python
X_train
```

```python
# from sklearn.model_selection import train_test_split
# X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)



from sklearn.svm import SVR
```

```python
svm_reg = SVR()
svm_reg.fit(X_train, y_train)
```

```python
y_pred_svm = svm_reg.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
np.sqrt(mean_squared_error(y_test, y_pred_svm))
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred_svm)
```
